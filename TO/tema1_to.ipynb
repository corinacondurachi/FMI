#Reprezentarea grafică a diferitelor tipuri de programe liniare



Un program liniar poate fi:


1.   Incompatibil (fără soluție)
2.   Compatibil cu o soluție optimă (de valoare finită)
3.   Compatibil cu  infinitate de soluții optime (de aceeași valoare)
4.   Compatibil cu optim infinit (nemărginită)


## 1.Sistem Incompatibil

max x + y  x + 2y >= 3, x + y <= 3, y <= -4

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.linspace(-5, 5, 200)

plt.figure(dpi=100)

plt.xlim(-5, 5)
plt.ylim(-5, 5)

# x + 2y >= 2
y = (2 - x)/2
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# x + y <= 3
y = x - 3
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# y <= -4
y = x * 0 - 4
plt.plot(x, y)
plt.fill_between(x, y, y - 10, alpha=0.2)

plt.show()
```

Dupa cum se vede in imagine, cele trei semiplane nu se intersecteaza simultan, deci sistemul este incompatibil.

## 2.Sistem Compatibil cu solutie optima de valoare finita

max x + yx + 2y >= 3, x - y >= 3

```
x = np.linspace(-5, 5, 200)

plt.figure(dpi=100)

plt.xlim(-5, 5)
plt.ylim(-5, 5)

# x + 2y >= 2
y = (2 - x)/2
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# x - 2y >= 3
y = (x - 3)/2
plt.plot(x, y)
plt.fill_between(x, y, y + 10, alpha=0.2)

# Punct de optim
plt.scatter(5/2, -1/4, c='red', zorder=5)

plt.show()
```

Cea mai mare valoare a lui $x + y$ se poate obține în punctul de intersecție ale dreptelor: x = 5/2, y = -1/4, x+y = 9/4



## 3.Sistem Compatibil cu optim unic, infinit

max x + y x + y <= 7




```
x = np.linspace(-7, 7, 200)

plt.figure(dpi=100)

plt.xlim(-7, 7)
plt.ylim(-7, 7)


# x + y <= 7
y = 7 - x
plt.plot(x, y)
plt.fill_between(x, y, y - 25, alpha=0.2)

plt.show()
```

Observăm că toate punctele de pe dreapta $x + y = 7$ maximizează funcția obiectiv $x + y$, deci avem o infinitate de soluții optime.
