```python
import random
import numpy as np
```

```python
#Metoda foloseste distributii Gamma

def generate_beta1(a, b):
    
    X1 = random.gammavariate(a, 1)
    X2 = random.gammavariate(b, 1)
    
    X = X1 / (X1 + X2)
    
    return X
```

```python
#Metoda foloseste distributii uniforme

def generate_beta2(a,b):
    
    while True:  
      U1 = random.uniform(0,1)
      U2 = random.uniform(0,1)

      # calculam V si T 
      V = U1 ** (1/a)
      T = U2 ** (1/(b-1))

      if V + T < 1:
          return V
```

```python
a = 0.3
b = 5
numar_exemple = 100000

set1 = [generate_beta1(a, b) for i in range(numar_exemple)]
set2 = [generate_beta2(a, b) for i in range(numar_exemple)]

media = a / (a + b)
standard_dev = np.sqrt((a * b)/(((a + b)**2)*(a + b + 1)))

print("Media si dispersia teoretice ar fi:")
print("Media: ", media)
print("Dispersia ", standard_dev)
print()

print("Media si dispersia obtinute pentru fiecare caz \n")
print("Media pentru prima metoda:", np.mean(set1))
print("Deviație standard pentru prima metoda:", np.std(set1))
print()
print("Media pentru metoda a doua:", np.mean(set2)) 
print("Deviație standard pentru metoda a doua:", np.std(set2))
```

```python
# Variabila normala prin metoda polara

def generate_normal_var():
    
    while True:
        # generam variabile independente U1 si U2
        U1 = random.uniform(0,1)
        U2 = random.uniform(0,1)

        V1 = 2 * U1 - 1
        V2 = 2 * U2 - 1
        S = V1**2 + V2**2
        
        if S < 1:

            Z1 = V1 * np.sqrt(-2*np.log(S)/S)
            Z2 = V2 * np.sqrt(-2*np.log(S)/S)
            return [Z1, Z2]
    
```

```python
numar_exemple = 100000
set1 = []

for i in range(numar_exemple):
    Z1, Z2 =  generate_normal_var()
    set1.append(Z1)
    set1.append(Z2)
    
print("Media si dispersia obtinute: \n")
print("Media:", np.mean(set1))
print("Deviație standard:", np.std(set1))
```
